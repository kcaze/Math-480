# Generating Random SYTs
Implement `random_SYT` and `random_SYT_2` and compare the efficacy on the shape (2,2,2).

Print out how long it takes to generate 10000 samples using each function and compare the distributions to the uniform distribution.

Plot the distributions using matplotlib and save them in the `figures` subfolder.

```python
shape = (2,2,2)

distribution = {str(syt): 0 for syt in SYT.SYTs(shape)}
start_time = time.time()
for i in range(10000):
  distribution[str(SYT.random_SYT(shape))] += 1
print(f"Time taken for method 1: {time.time() - start_time}")

distribution2 = {str(syt): 0 for syt in SYT.SYTs(shape)}
start_time = time.time()
for i in range(10000):
  distribution2[str(SYT.random_SYT_2(shape))] += 1
print(f"Time taken for method 2: {time.time() - start_time}")

plt.bar(distribution.keys(), distribution.values())
plt.xticks(rotation=90)
plt.title("Distribution of SYTs (Algorithm 1)")
plt.show()
plt.savefig("figures/distribution.png")

plt.bar(distribution2.keys(), distribution2.values())
plt.xticks(rotation=90)
plt.title("Distribution of SYTs (Algorithm 2)")
plt.show()
plt.savefig("figures/distribution2.png")
```
