## Imports
The files you need for this notebook are imported in the cell below.

```python
import catalan
import os
import time
import matplotlib.pyplot as plt
dirname = os.path.dirname("./")
```

## Generate Parenthesizations

Generate all possible parenthesizations of length $n$ for $1 \le n \le 10$. Files should be outputted to `data/catalan_parenthesizations_n.txt` and a plot of the time used to generate each data file should go in `figures/catalan_parenthesizations.png`.

```python
import time
timing_results = []

data_dir = 'data'
figures_dir = 'figures'

os.makedirs(data_dir, exist_ok=True)
os.makedirs(figures_dir, exist_ok=True)

for n in range(1, 11):
    start_time = time.time()
    parenthesizations = catalan.parenthesizations(n)
    end_time = time.time()
    duration = end_time - start_time
    timing_results.append((n, duration))

    # Save parenthesizations to file
    with open(os.path.join(data_dir, f'catalan_parenthesizations_{n}.txt'), 'w') as file:
        for p in parenthesizations:
            file.write(f'{p}\n')

# Plotting the timing results
ns = [result[0] for result in timing_results]
durations = [result[1] for result in timing_results]

plt.figure(figsize=(10, 6))
plt.plot(ns, durations, marker='o', linestyle='-', color='b')
plt.title('Time to Generate Parenthesizations')
plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.savefig(os.path.join(figures_dir, 'catalan_parenthesizations.png'))
plt.show()
```
